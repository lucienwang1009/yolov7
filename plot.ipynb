```python
scales = [80, 40, 20]

def get_bbox(targets, layer_idx, idx=0):
    left, top = int(scales[layer_idx] * (targets[idx][2] - targets[idx][4] / 2)), \
        int(scales[layer_idx] * (targets[idx][3] - targets[idx][5] / 2))
    right, bottom = int(scales[layer_idx] * (targets[idx][2] + targets[idx][4] / 2)), \
        int(scales[layer_idx] * (targets[idx][3] + targets[idx][5] / 2))
    return left, top, right, bottom

def fuse(query_features, support_features_cls):
    # support_ft_cls = list(layer[0] for layer in support_ft)

    fused_features = []
    agg_features = []
    att_weights = []
    for layer_idx in range(len(query_features)):
        fused_feature = []
        feature = query_features[layer_idx]
        B, C, H, W = feature.size()
        # query = feature.permute(0, 2, 3, 1).reshape(B, H * W, -1)
        query = model.qs[layer_idx](
            feature.permute(0, 2, 3, 1).reshape(B, H * W, -1)
        )
        query = query - torch.mean(query, dim=1, keepdim=True)

        for support_feature in support_features_cls[layer_idx]:
            _, _, s_H, s_W = support_feature.shape
            # NHWC
            support_feature_flatten = support_feature.permute(0, 2, 3, 1) \
                .reshape(1, s_H * s_W, -1)
            key = model.ks[layer_idx](support_feature_flatten)
            # key = support_feature_flatten
            key -= torch.mean(key, dim=1, keepdim=True)
            
            attention_weight = torch.matmul(
                query, key.transpose(1, 2)) / math.sqrt(C)
            # attention_weight_sig = F.sigmoid(attention_weight)
            attention_weight = F.softmax(attention_weight / 1, dim=2)
            # attention_weight[attention_weight_sig < 0.6] = 0
            
            # left, top, right, bottom = get_bbox(targets, layer_idx) 
            # att_weights.append(
            #     attention_weight.reshape(B, H, W, -1)[
            #         0, top:bottom, left:right
            #     ].reshape(-1, s_H, s_W).mean(dim=0)
            # )
            
            fused_f = torch.matmul(attention_weight, support_feature_flatten)
            # fused_f = fused_f / torch.sqrt(torch.sum(fused_f ** 2, -1, keepdim=True))
            fused_feature.append(
                fused_f.reshape(B, H, W, C).permute(0, 3, 1, 2)
            )
        agg_feature = torch.stack(fused_feature, dim=0).mean(dim=0)
        agg_features.append(agg_feature)
        
        fused = model.fuses[layer_idx](torch.concat(
                      [feature, agg_feature], dim=1))
        # fused = feature * agg_feature
        fused_features.append(
            # fused
            F.cosine_similarity(agg_feature.permute(0, 2, 3, 1).reshape(B * H * W, -1), 
                                feature.permute(0, 2, 3, 1).reshape(B * H * W, -1), dim=-1).reshape(B, H, W, -1).permute(0, 3, 1, 2)
            # ((fused - feature) ** 2).mean(dim=1, keepdim=True)
        )
    return fused_features, agg_features, att_weights
```
